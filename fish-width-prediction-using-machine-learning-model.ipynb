```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
```

```python
#load the dataset
data=pd.read_csv('/kaggle/input/fish-dataset/Fish.csv')
data.head(10)
```

```python
#List the name of columns
data.columns
```

```python
#check if there is any missig value
data.info()
```

```python
#input variables
x=data[['Weight', 'Length1', 'Length2', 'Length3', 'Height']]
x.head(4)
```

```python
#output variables
y=data[['Width']]
y.head(4)
```

```python
#split the data to 80% for trainting and 20% for testing
x_train, x_test, y_train, y_test = train_test_split(x, y,random_state=1, test_size=0.2)
```

```python
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

# linear regression

```python
#linear regression 
regressor = LinearRegression()
```

```python
# Fit the linear regression model to the training data
regressor.fit(x_train_scaled, y_train)
```

```python
# Evaluating the model quality
r_sq = regressor.score(x_test_scaled, y_test)
print('score of r square: ', r_sq)
```

# Random Forest
